typical customer profile.

more on average?

## Identify High-Value Customers
Filter customers with a Spending Score > 80 and calculate their average Annual
Income.

Identify the top 10 customers by Spending Score. What do they have in common
(e.g., age group, gender)?

```python
def categorise_spending(score):
    if score > 80: return 'A'
    elif score > 70: return 'B'
    elif score > 60: return 'C'
    else: return 'D'

df2 = df

df2['spending_grade'] = df2['Spending Score (1-100)'].apply(categorise_spending)

df2
```
